```python
df3=df.copy()
df3=df3.dropna()
print(df3)
```

```python
df3.dropna(inplace=True , axis=1)#inplace says not to reflect the null values  change in the df
df3
```

```python
df3.dropna(inplace=True,how='all')#drops both rows and column that are not null
df3
```

```python
mv=df3['SEM1'].mean()#it will replace the null value with mean value
df3=df3.fillna(mv)
print(df3)
```

```python
df3=df3.drop_duplicates()
df3
```

```python
df3.rename(columns={'Grade':'GPA'},inplace=True)
df3.head()
```

```python
df3.columns=['Student_name','Mr/Ms','1stsem','2ndsem','3rdsem','F Outcome','Placed?']
df3.head()
```

```python
df3['Avq score']=(df3['1stsem']+df3['2ndsem']+df3['3rdsem'])/3
df3.head()
```
